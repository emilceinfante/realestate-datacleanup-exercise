#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import matplotlib.pyplot as plt

# Definir las poblaciones que queremos filtrar
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame usando la función isin
subconjunto = ds[ds['level5'].isin(poblaciones)]

# Normalizar los precios para cada población
subconjunto['normalized_price'] = subconjunto.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

# Trazar los histogramas
plt.figure(figsize=(10, 6))
for poblacion in poblaciones:
    plt.hist(subconjunto[subconjunto['level5'] == poblacion]['normalized_price'], 
             bins=10, 
             alpha=0.5, 
             label=poblacion)

plt.title('Histogramas de Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y')
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
#Definir las poblaciones que queremos filtrar
poblaciones = ["Getafe", "Alcorcón"]

# Filtrar el DataFrame usando la función isin
subconjunto = ds[ds['level5'].isin(poblaciones)]

# Calcular el precio por metro cuadrado
subconjunto['price_per_sqm'] = subconjunto['price'] / subconjunto['surface']

# Calcular la media del precio por metro cuadrado por población
media_precios = subconjunto.groupby('level5')['price_per_sqm'].mean()

# Imprimir los resultados
print("Precio promedio por metro cuadrado:")
print(media_precios)

# Conclusión
if media_precios['Getafe'] > media_precios['Alcorcón']:
    conclusion = "El precio por metro cuadrado en Getafe es mayor que en Alcorcón."
elif media_precios['Getafe'] < media_precios['Alcorcón']:
    conclusion = "El precio por metro cuadrado en Alcorcón es mayor que en Getafe."
else:
    conclusion = "Los precios por metro cuadrado en Getafe y Alcorcón son iguales."
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
import matplotlib.pyplot as plt
from ipyleaflet import Map, basemaps, CircleMarker

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
subconjunto = ds[ds['level5'].isin(poblaciones)]

# Crear un diccionario para almacenar las coordenadas
coordenadas = {poblacion: {'latitud': [], 'longitud': []} for poblacion in poblaciones}

# Llenar el diccionario con las coordenadas
for index, row in subconjunto.iterrows():
    poblacion = row['level5']
    coordenadas[poblacion]['latitud'].append(row['latitude'])
    coordenadas[poblacion]['longitud'].append(row['longitude'])

map = Map(center=(40.35, -3.8), zoom=11.5, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)

colors = {
    "Fuenlabrada": "blue",
    "Leganés": "green",
    "Getafe": "purple",
    "Alcorcón": "red"
}

# Trazar las coordenadas de las poblaciones
for poblacion in poblaciones:
    latitudes = coordenadas[poblacion]['latitud']
    longitudes = coordenadas[poblacion]['longitud']
    for lat, lon in zip(latitudes, longitudes):
        marker = CircleMarker(
            location = (lat, lon), 
            radius = 5,
            color = colors[poblacion],
            fill_color = colors[poblacion], 
            fill_opacity=0.8

        )
        map.add_layer(marker)
map



```
